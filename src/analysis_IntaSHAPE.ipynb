```python
%matplotlib inline
import seaborn as sns
import pandas as pd


df_NOSHAPE = pd.read_csv(heatmap_csv,sep=';')
df_SHAPE = pd.read_csv('SHAPE-'+heatmap_csv,sep=';')

exon_regions = [(1,387), (244,387), (294,549), (811,880), (815,880), (1017,1410), (1511,2124), (2236,2621)]

from matplotlib import pyplot as plt
def plot_minE_heatmap(df, figtitle, fig=None, ax=None):
    if (fig is None):
        fig, ax = plt.subplots(figsize=[50,5])
    mydf = df.iloc[:,1:]#.transpose()
    # mydf[mydf>0] = 'NaN'
    mydf[mydf>0] = None
    [ax.axvspan(ex[0]-1,ex[1]-1, alpha=0.2) for ex in exon_regions]

    sns.heatmap(mydf.transpose(),ax=ax)
    ax.set_ylim(ax.get_ylim()[1],ax.get_ylim()[0])

    import matplotlib.ticker as plticker
    loc = plticker.MultipleLocator(base=100,) # this locator puts ticks at regular intervals
    ax.xaxis.set_major_locator(loc)
    majorFormatter = plticker.FormatStrFormatter('%d')
    ax.xaxis.set_major_formatter(majorFormatter)
    ax.set_title(figtitle)
    fig.savefig(figtitle+'.png')
    # sns.heatmap(df_SHAPE.iloc[:,1:5],ax=ax)

myfig, (ax1,ax2) = plt.subplots(figsize=[50,10],nrows=2)
plot_minE_heatmap(df_NOSHAPE, target_gene_name +'-'+query_id+ '-NOSHAPE', fig=myfig, ax=ax1)
plot_minE_heatmap(df_SHAPE, target_gene_name + '-'+query_id+'-SHAPE', fig=myfig, ax=ax2)
```
